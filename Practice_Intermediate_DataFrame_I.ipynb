## Sorting in DataFrame

```python
data_1 = {'Name':['Budi', 'Andi', 'Lala'],
          'Age':[20,25,26],
          'Fruit':['Strawberry','Orange','Apple']
    }

df_1 = pd.DataFrame(data_1)
df_1
```

```python
df_1.sort_values(by=['Name'])
```

```python
df_1.sort_values(by=['Fruit'], ascending=False)
```

```python
df_1.sort_values(by=['Name', 'Fruit'])
```

## Creating Additional Column

```python
#additional column with value from other column
data_0['year_birth'] = 2021 - data_0['Age']

data_0
```

```python
data_0['New_Fare'] = data_0['Fare'] * 1.1
data_0
```

```python
data_0['random'] = data_0['Survived']
data_0
```

```python
# menyisipkan kolom di posisi tertentu
data_0.insert(loc=3, column = 'Test', value = data_0['PassengerId']*2)
```

```python
data_0
```

```python
data_0.columns.get_loc('Age')
```

```python
# additional column with single value
data_0['random'] = 0
data_0
```

```python
# additional column with Value from other column (categorized)
# menggunakan NumPy

condition_pclass = [data_0['Pclass'] == 1,
                    data_0['Pclass'] == 2,
                    data_0['Pclass'] == 3]

value_pclass = ['VIP Class', 'Business Class', 'Economu Class']
data_0['Category'] = np.select(condition_pclass,value_pclass)
data_0
```

```python
data_0.drop(columns= [ 'Test', 'random','New_Fare'], inplace=True)
data_0
```

## Grouping and Aggregate in DataFrame

```python
data_gender = data_0.groupby(by=['Sex'])[['PassengerId']].count()
data_gender
```

```python
data_gender['PassengerId'] = (data_gender['PassengerId'] / data_gender['PassengerId'].sum()*100)
data_gender
```

```python
data_0.pivot_table(values= 'Age', columns=['Sex', 'Category'], aggfunc='mean')
```

```python
data_0.pivot_table(values= 'Fare', columns=['Sex', 'Category'], aggfunc='max')
```
